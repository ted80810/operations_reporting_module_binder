### Break up rows with store information into seperate DataFrame

The code below breaks up each store into a dataframe. In order to do this you must specify the store number of the desired store, the starting row of each unique position and the end row of each unique position. The function will break up each store position into a dataframe. Lastly, the code will concatenate each dataframe into a single dataframe and conver it into a csv.

Store #1

```python
store_number = '01'
```

```python
start_rows = [4,135,172,179,189,204] # <----change to start row of position
end_rows = [131,168,175,185,200,210] # <----change to end row of position

position_list = ['crew','crew_trainer','maintenance','star','summer_student', 'trailer']
```

```python
def store_df(hr, store_number, start_rows, end_rows, position_list):
    for i in range(len(start_rows)):
        for position in position_list:
            position_df = hr.iloc[start_rows[i]:end_rows[i]]
            position_df.iat[0,1] = position
            position_df.iat[0,2] = store_number
            position_df = position_df.ffill(axis = 0)
    return position_df
```

```python
store_01 = store_df(hr, store_number, start_rows, end_rows, position_list)
```
